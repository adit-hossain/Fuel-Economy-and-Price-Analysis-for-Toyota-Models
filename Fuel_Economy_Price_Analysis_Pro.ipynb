# Fuel Economy & Price Analysis of Toyota Models

This notebook analyzes trends in fuel efficiency and price for Toyota models.

## Steps:
1. Load dataset
2. Visualize distributions and relationships
3. Perform regression analysis
4. Present visual results


```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Example dataset
data = {
    'Model': ['Corolla', 'Camry', 'RAV4', 'Prius', 'Yaris', 'Highlander', 'Tacoma', 'Supra'],
    'Mileage_km': [50000, 60000, 45000, 30000, 70000, 55000, 80000, 20000],
    'Fuel_Economy_kmpl': [15, 13, 12, 20, 18, 10, 8, 9],
    'Price_USD': [15000, 18000, 22000, 25000, 12000, 30000, 27000, 50000]
}

df = pd.DataFrame(data)

# Overview
print(df.describe())

# Mileage distribution
sns.histplot(df['Mileage_km'], kde=True, color='skyblue')
plt.title("Mileage Distribution")
plt.show()

# Fuel economy vs price
sns.scatterplot(x='Fuel_Economy_kmpl', y='Price_USD', hue='Model', data=df, s=100, palette="viridis")
plt.title("Fuel Economy vs Price")
plt.show()

# Mileage vs price regression
X = df[['Mileage_km']]
y = df['Price_USD']

model = LinearRegression()
model.fit(X, y)

df['Predicted_Price'] = model.predict(X)

sns.scatterplot(x='Mileage_km', y='Price_USD', data=df, label='Actual')
sns.lineplot(x='Mileage_km', y='Predicted_Price', data=df, color='red', label='Regression Line')
plt.title("Mileage vs Price Regression")
plt.show()

print(f"Regression Coefficient: {model.coef_[0]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")

# Correlation heatmap
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
